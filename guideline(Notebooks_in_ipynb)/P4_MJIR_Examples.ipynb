````python
/* Bubble sort code */

class Main {
    public static void main(String[] args) {
    	int[] v = new int[100];
    	int n, c, d, swap;
    	// Number of elements:
    	n = 25;
    	for (c = 0; c < n; c = c+1)
            v[c] = n - c;
    	for (c = 0; c < n-1; c = c+1)
            for (d = 0; d < n-c-1; d = d+1)
                if (v[d] > v[d+1]) {
                    swap = v[d];
                    v[d] = v[d+1];
                    v[d+1] = swap;
                }
    	print("Sorted list in ascending order: ");
    	for (c = 0; c < n; c = c+1)
            print(v[c], " ");
    }
}
````
